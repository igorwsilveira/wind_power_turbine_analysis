```python
# Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

### 1) Read CSV File

```python
turbine = pd.read_csv('T1.csv')
turbine.columns = ['Date/Time', 'ActivePower(kW)', 'WindSpeed(m/s)', 'TheoricalCurve(KWh)', 'WindDirection(ยบ)']
del turbine['WindDirection(ยบ)']
turbine['Date/Time'] = pd.to_datetime(turbine['Date/Time'])
display(turbine)
```

### 2) Data Visualization (Wind Speed vs Active Power)

```python
sns.scatterplot(data=turbine, x='WindSpeed(m/s)', y='ActivePower(kW)')
sns.set(rc={"figure.figsize":(8, 4)})
```

### 3) Theorical Curve Visualization

```python
sns.scatterplot(data=turbine, x='WindSpeed(m/s)', y='TheoricalCurve(KWh)')
sns.set(rc={"figure.figsize":(8, 4)})
```

### 4) Limit of Performance

The Limit of Performance is based on the 5% up or down. This way we can make sure the turbine is working properly with an extra performance, generating power (up to 5%) or has any issue with the engine or any other components that may require attention or even repair (down to 5%).

```python
real_power = turbine['ActivePower(kW)'].tolist()
theorical_power = turbine['TheoricalCurve(KWh)'].tolist()
max_power = []
low_power = []
limit = []
for power in theorical_power:
    max_power.append(power*1.05)
    low_power.append(power*0.95)

for p, power in enumerate(real_power):
    if power>=low_power[p] and power<=max_power[p]:
        limit.append('Ok')
    elif power == 0:
        limit.append('Neutral')
    else:
        limit.append('Out')

print(limit.count('Ok')/len(limit))

    
```

### 5) Adding Column "Threshold" in the DataFrame

To do this I will create an extra column in the data frame so can measure and compared the actual power versus the theoretical curve.

```python
turbine['Threshold'] = limit
display(turbine)
```

### 6) Plotting Graphics

This graphic will be helpful for the maintenance crew or the engineer responsible for the turbines in the sector.
Giving an overview about the general condition of the turbine site, this way they can visualize the outliers and raise questions about the performance, issues, and improvements that need to be made in comparison to others.

```python
sns.scatterplot(data=turbine, x='WindSpeed(m/s)', y='ActivePower(kW)', hue='Threshold', palette="magma", s=3)
sns.set(rc={"figure.figsize":(8, 4)})
```
